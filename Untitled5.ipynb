```python
train_data = data.drop(['Name', 'Fare', 'Embarked', 'Cabin', 'Ticket', 'PassengerId'], axis=1)
```

```python
#Проверим наличие незаполненных данных о возрасте
train_data.count()
```

```python
#Как было видно ранее, медианный возраст пассажиров меняется в зависимости от класса каюты
#Также это значение разное у мужчин и женщин, поэтому заполним недостающие значение средним возрастом по классу в зависимости от пола
arr = pd.unique(train_data['Pclass'])
```

```python
i=0
while i < 3:
    MeanOfAgeMen = train_data[(train_data['Pclass'] == arr[i]) & (train_data['Sex'] == 'male') & (~pd.isna(train_data['Age']))]['Age'].mean()
    train_data[(train_data['Pclass'] == arr[i]) & (train_data['Sex'] == 'male')] = train_data[(train_data['Pclass'] == arr[i]) & (train_data['Sex'] == 'male')].fillna(MeanOfAgeMen)
    MeanOfAgeWomen = train_data[(train_data['Pclass'] == arr[i]) & (train_data['Sex'] == 'female') & (~pd.isna(train_data['Age']))]['Age'].mean()
    train_data[(train_data['Pclass'] == arr[i]) & (train_data['Sex'] == 'female')] = train_data[(train_data['Pclass'] == arr[i]) & (train_data['Sex'] == 'female')].fillna(MeanOfAgeWomen)
    i+=1
```

```python
#Проверяем работу заполнения средними
train_data.count()

train_data['Survived'] = pd.factorize(train_data['Survived'])[0]
labels_data = train_data['Survived']
```

```python
#Превратим категориальный признак в количественный
train_data['Sex'] = pd.factorize(train_data['Sex'])[0]
train_data = train_data.drop('Survived', axis=1)
```

```python
#Масштабируем наши данные для корректной работы t-SNE 
scaler = StandardScaler()
train_data_scaled = scaler.fit_transform(train_data)
```

```python
tsne = TSNE(random_state=42)
tsne_representation = tsne.fit_transform(train_data_scaled)
```

```python
plt.scatter(tsne_representation[:, 0], tsne_representation[:, 1], 
            c=data['Survived'].map({True: 'blue', False: 'orange'}));
```

```python
#Разделим наши данные в пропорции 80/20
X_train, X_holdout, y_train, y_holdout = train_test_split(train_data_scaled, labels_data, test_size=0.2,
random_state=42)
```

```python
cv = StratifiedKFold(labels_data, n_folds=5, shuffle=True, random_state=42)
```

```python
#Получим значения параметров дерева для наилучшего результата
tree = DecisionTreeClassifier()
tree_params = {'max_depth': range(5,10),
'max_features': range(1,5)}
```

```python
tree_grid = GridSearchCV(tree, tree_params,
cv=5, n_jobs=-1,
verbose=True)
```

```python
tree_grid.fit(X_train, y_train)
```

```python
tree_grid.best_score_
```

```python
tree_grid.best_params_
```

```python
tree = DecisionTreeClassifier(max_depth=6, max_features=4, random_state=42)
scores_tree = cross_val_score(tree, train_data, labels_data, cv=cv, n_jobs=-1)
scores_tree.mean()
```

```python
knn_pipe = Pipeline([('scaler', StandardScaler()), ('knn', KNeighborsClassifier(n_jobs=-1))])
```

```python
#Тем же способом подберем наилучучшее значения параметра
knn_params = {'knn__n_neighbors': range(1, 15)}

knn_grid = GridSearchCV(knn_pipe, knn_params,
cv=5, n_jobs=-1,
verbose=True)

knn_grid.fit(X_train, y_train)
```

```python
knn_grid.best_params_, knn_grid.best_score_
```
